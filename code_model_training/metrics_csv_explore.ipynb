```python
import pandas as pd

# Path to your metrics file
csv_path = "/home/puneet/maninder/code_model_training/modelResults/her2st_stnet_224/result_vit_2025-11-12_00-56-07/metrics_result.csv"

# Load the CSV
df = pd.read_csv(csv_path)

# Filter for validation phase only
val_df = df[df["phase"] == "val"]

# Metrics of interest
metrics = [
    "pearson_mean_cell_wise", "pearson_mean_genewise",
    "spearman_mean_cell_wise", "spearman_mean_genewise",
    "l1_error_mean", "l2_errors_mean"
]

# Compute per-fold mean and std
fold_stats = val_df.groupby("fold")[metrics].agg(["mean", "std"])

# Compute overall mean and std across all folds
overall_stats = {
    metric: {
        "overall_mean": val_df[metric].mean(),
        "overall_std": val_df[metric].std()
    } for metric in metrics
}

# Convert to DataFrame
overall_stats_df = pd.DataFrame(overall_stats).T

print("\n=== Per-Fold Mean & Std ===")
print(fold_stats)

print("\n=== Overall Mean & Std ===")
print(overall_stats_df)
```

```python
import pandas as pd

# Path to your metrics file
csv_path = "/home/puneet/maninder/code_model_training/modelResults/her2st_stnet_pretrained_true_amt_224/result_vit_2025-11-14_16-48-59/metrics_result.csv"

# Load the CSV
df = pd.read_csv(csv_path)

# Filter for validation phase only
val_df = df[df["phase"] == "val"]

# Metrics of interest
metrics = [
    "pearson_mean_cell_wise", "pearson_mean_genewise",
    "spearman_mean_cell_wise", "spearman_mean_genewise",
    "l1_error_mean", "l2_errors_mean", "mape_mean"
]

# --- Step 1: Find the row with max spearman_mean_genewise per fold ---
best_rows = val_df.loc[val_df.groupby("fold")["spearman_mean_genewise"].idxmax()]

# --- Step 2: Compute mean and std across these best rows ---
best_stats = best_rows[metrics].agg(["mean", "std"])

print("\n=== Best (Max Spearman Gene-wise) Row per Fold ===")
print(best_rows[["fold", "epoch", "spearman_mean_genewise"] + metrics])

print("\n=== Mean & Std Across Best Folds ===")
print(best_stats)
```
